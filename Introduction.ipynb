```python
def gcd_two(m,n):
  lm = []
  for i in range(1,m+1):
    if (m%i)==0:
      lm.append(i)
  ln=[]   
  for i in range(1,n+1):
    if (n%i)==0:
      ln.append(i)   

  cf=[]
  for n in lm:
    if n in ln:
      cf.append(n)   

  return max(cf)          
```

## Improving Naive GCD

```python
# improving naive gcd

def gcd_non_naive(m,n):
  ls = []
  for i in range(1,min(m,n)+1):   # because factors never cross the min of both values
    if m%i == 0 and n%i == 0:
      ls.append(i)

  return max(ls)   #return ls[-1] can be used  
```

```python
gcd_non_naive(14,63)
```

```python
# gcd   with no list

def gcd_no_list(m,n):
  for i in range(1,min(m,n)+1):   # because factors never cross the min of both values
    if m%i == 0 and n%i == 0:
      mrcf = i   # most recent common factor
  return mrcf
```

```python
gcd_no_list(14,63)
```

```python
def gcd_while(m,n):
  i = min(m,n)

  while i>0:
    if (m%i) == 0 and (n%i)==0:
      return i
    else:
      i = i-1  
```

```python
gcd_while(14,63)
```

## Euclids algorithm for GCD


```python
def gcd(m,n):
#m>n then d # Assume m>n
  if m<n:
    (m,n)=(n,m)

  if (m%n) == 0:
    return n
  else:
    diff = m-n
    #diff > n ? possible
  return gcd(max(n,diff),min(n,diff))       # recursion
```

```python
gcd(14,63)
```

```python
def gcd_original(m,n):
#m>n then d # Assume m>n
  if m<n:
    (m,n)=(n,m)

  if (m%n) == 0:
    return n
  else:
    return gcd_original(n,m%n)
    #diff > n ? possible  
```

```python
gcd_original(14,63)
```
